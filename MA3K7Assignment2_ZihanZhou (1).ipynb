### Simulate the game
The `game` function below simulates the behaviour of my friend by generating a random response. It is designed based on the assumption that my friend has no intelligence.

```python
n = 3
matrix = np.array([['.']*n]*n)

print('Matrix: \n', matrix)

def check(x,y, matrix):
    global n
    while True:
        if not ((1<=x<=n) and (1<=y<=n) and matrix[x-1, y-1] == '.') : 
            print("Nope... choose again")
            x = int(input(f"Your turn: on which row would you like to put a zero (up to {n})?"))
            y = int(input(f"Your turn: on which column would you like to put a zero (up to {n})?"))
        else: return x-1, y-1
    
def game(myturn):
    global matrix
    global n
    if myturn == 1:
        x = int(input(f"Your turn: on which row would you like to put a zero (up to {n})?"))
        y = int(input(f"Your turn: on which column would you like to put a zero (up to {n})?"))
        x,y = check(x,y, matrix)
        matrix[x, y] = 0
        print('Matrix: \n', matrix)
        print('------------------------')
        if '.' in matrix:
            game(0)
        else:
            if LA.det(matrix) == 0:
                print('You win!')
                return 0
            else:
                print('You lose!')
                return 0
    
    else:
        coords = np.array([])
        for i in range(n):
            for j in range(n):
                if matrix[i, j] == '.':
                    new = np.array([i, j])
                    if np.size(coords) == 0:
                        coords = np.hstack((coords, new))
                    else:
                        coords = np.vstack((coords, new))
        
        if np.size(coords) == 2:
            coord = coords
        else:
            rng = np.random.default_rng()
            coord = rng.choice(coords)
        x = int(coord[0])
        y = int(coord[1])
        print(f"My friend's turn: put 1 at row{x+1}, column{y+1}")
        matrix[x, y] = 1
        print('Matrix: \n', matrix)
        print('------------------------')
        if '.' in matrix:
            game(1)
        else:
            if LA.det(matrix) == 0:
                print('You win!')
                return 0
            else:
                print('You lose!')
                return 0
game(0)
```

### Calculate the number of possible patterns of the matrix  
For an $n \times n$ matrix game, there are $n^2$ possible positions to put our 0s and 1s.  
- If $n$ is even, there are $n^2/2$ 0s and 1s repsectively, and we have $$\binom{n^2}{n^2/2}$$ possible patterns.
- If $n$ is odd, there are $(n^2-1)/2$ 0s and $(n^2+1)/2$ 1s repsectively(or the other way round, dependeing on who's starting first), and we have $$\binom{n^2}{(n^2-1)/2}$$ possible patterns.

```python
import math
ns = np.arange(4, 11)
for n in ns:
    if n%2 == 0:
        m = n*n
        k = int(n*n/2)
        num = math.comb(m, k)
        
    if n%2 == 1:
        m = n*n
        k = int((n*n-1)/2)
        num = math.comb(m, k)
    
    print(f"n = {n}, number of possible patterns: {num}")
```

### Is this a fair game?
Generate the probability of getting zero determinant for each $n$ if both players play randomly if played 100000 times.

```python
def generate(dim, order):

    rng = np.random.default_rng()
    sample = 1000
    wins = 0
    
    if order == 1: # I start first

        if dim%2 == 0:
            elements = [1]*int(dim**2 / 2) + [0]* int(dim**2 / 2)
        
    
            for i in np.arange(sample):
                A=rng.choice(elements, (dim,dim))
                detA = LA.det(A)
            
                if detA == 0:
                    wins += 1
            
        if dim%2 == 1:
            elements = [1]* int((dim**2 - 1)/2) + [0]* int((dim**2 + 1)/2)
        
            for i in np.arange(sample):
                A = rng.choice(elements, (dim, dim))
                detA = LA.det(A)
                
                if detA == 0:
                    wins += 1
    
    if order == 0: # My friend starts first
    
        if dim%2 == 0:
            elements = [1]*int(dim**2 / 2) + [0]* int(dim**2 / 2)
        
    
            for i in np.arange(sample):
                A=rng.choice(elements, (dim,dim))
                detA = LA.det(A)
                if detA == 0:
                    wins += 1
            
        if dim%2 == 1:
            elements = [0]* int((dim**2 - 1)/2) + [1]* int((dim**2 + 1)/2)
        
            for i in np.arange(sample):
                A = rng.choice(elements, (dim, dim))
                detA = LA.det(A)
                if detA == 0:
                    wins += 1
    chance = wins/sample
    return chance
    

```

```python
chances1 = []
chances0 = []
for n in ns:
    chance = generate(n, 1)
    chances1.append(chance*100)
    print(f"n = {n}, I start first, chance of winning is {chance*100}%. ")
    chance = generate(n, 0)
    chances0.append(chance*100)
    print(f"n = {n}, my friend starts first, chance of winning is {chance*100}%. ")
```
